```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Set a nice style for the plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Generate synthetic data with a clear quadratic relationship
np.random.seed(42)
X_nonlinear = np.linspace(-5, 5, 100).reshape(-1, 1)
# y = 0.5x^2 - 2x + 1 + noise
y_nonlinear = 0.5 * (X_nonlinear.ravel()**2) - 2 * X_nonlinear.ravel() + 1 + np.random.normal(0, 3, 100)

# Plot the raw data
plt.scatter(X_nonlinear, y_nonlinear, alpha=0.7)
plt.title("Synthetic Non-Linear (Quadratic) Data")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.show()
```

```python
# Fit a simple linear regression model to the non-linear data
model_bad_fit = LinearRegression()
model_bad_fit.fit(X_nonlinear, y_nonlinear)
y_pred_bad_fit = model_bad_fit.predict(X_nonlinear)
```

```python
# Calculate residuals for the bad fit
residuals_bad_fit = y_nonlinear - y_pred_bad_fit

# Plot 1: Actual vs. Predicted
plt.subplot(1, 2, 1)
plt.scatter(y_pred_bad_fit, y_nonlinear, alpha=0.7)
plt.plot([y_nonlinear.min(), y_nonlinear.max()], [y_nonlinear.min(), y_nonlinear.max()], '--r', lw=2)
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.title("Actual vs. Predicted (Bad Fit)")

# Plot 2: Residuals vs. Predicted
plt.subplot(1, 2, 2)
sns.scatterplot(x=y_pred_bad_fit, y=residuals_bad_fit, alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs. Predicted (Bad Fit)")

plt.tight_layout()
plt.show()
```
